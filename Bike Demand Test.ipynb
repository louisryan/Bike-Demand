<img src="files/data/bikes.png">

Bike sharing systems are a means of renting bicycles where the process of obtaining membership, rental, and bike return is automated via a network of kiosk locations throughout a city. Using these systems, people are able rent a bike from a one location and return it to a different place on an as-needed basis. Currently, there are over 500 bike-sharing programs around the world.

The data generated by these systems makes them attractive for researchers because the duration of travel, departure location, arrival location, and time elapsed is explicitly recorded. Bike sharing systems therefore function as a sensor network, which can be used for studying mobility in a city. In this competition, participants are asked to combine historical usage patterns with weather data in order to forecast bike rental demand in the Capital Bikeshare program in Washington, D.C.

### Modelling

We have got a better understanding of the data, allowing us to bring this inferred knowledge into our modelling.

Of course there are going to be seasonal/weather effects in forecasting demand. I'm going to hypothesise that forecast will be at least partially due to changing daylight hours. We'll compute the number of hours of daylight and use this to de-trend the data:

```python
def hours_daylight(date, axis=23.44, latitude=47.61):
    
    # Compute the hours in a day given a date
    diff = date - pd.datetime(2000, 12, 21)
    day = diff.total_seconds() / 24. / 3600
    day %= 365.25
    m = 1. - np.tan(np.radians(latitude)) * np.tan(np.radians(axis) * np.cos(day * np.pi / 182.625))
    m = max(0, min(m, 2))
    return 24. * np.degrees(np.arccos(1 - m)) / 180.
    
#hours_daylight(my_date, 23.44, 47.61)
```

Next, we'll extract the year, month, day and hour from the datetime hour index and use these as additional features in the model.

```python
def split_datetime(data):
    data['year']  = data.index.year
    data['month'] = data.index.month
    data['day']   = data.index.day
    data['hour']  = data.index.hour
    #data = data.reset_index()
    return data

df = split_datetime(df)
df.head()
```

We want to extract all of the columns except for the ones specified below:

```python
features = df[[col for col in df.columns if col not in ['datetime', 'casual', 'registered', 'count']]].values
target = df[['count']].values

print 'The shape of the training features is: %s' % (features.shape,)
print 'The shape of the target feature is: %s' % (target.shape,)
```
